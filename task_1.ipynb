[문제 3]

사용자로부터 여러 개의 숫자를 입력받아 리스트에 저장한 후, 사용자가 'done'이라고 입력할 때까지 계속 입력받으세요. 

- 입력 중 숫자가 아닌 값이 입력되면 ValueError를 처리
- 숫자 외 값이 입력된 경우 어떤 값이 잘못 입력되었는지 사용자에게 알려주는 문구 출력<br>
  (예: f"잘못된 값: {변수}. 숫자를 입력해주세요.")



````python

def digit_set():
    숫자 = [] 

    while True:
        입력 = input("숫자를 입력하세요 ('done' 입력 시 종료): ")
        if 입력 == "done" :
            break
        try:
            숫자.extend([float(x) for x in 입력.split()])
        except ValueError:
            print(f"잘못된 값: {입력}. 숫자만 입력하세요.")

    print("입력한 숫자 리스트:", 숫자)
    return 숫자

digit_set()
````

[문제 7]

함수 실행 시간을 측정하는 데코레이터 timer_decorator를 작성하세요. 

임의의 함수가 실행되는 시간을 측정하여 "함수 실행 시간: x.xx초" 형태로 출력하세요.

````python
import time

def timer_decorator(function):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = function(*args, **kwargs)
        end = time.time()
        elapsed = end - start
        print(f"함수 실행 시간:{elapsed:.2f}초")
        return result
    return wrapper

@timer_decorator
def f():
    for i in range(54734579):
        pass
f()
````

````python
p1 = 3.141592

# 사용자로부터 반지름 값을 입력받는 함수입니다.
def number_input():
    # 사용자에게 반지름을 입력하라는 메시지를 출력하고 입력받습니다.
    value = input("반지름을 입력해주세요")
    # 입력받은 값을 float 타입으로 변환하여 반환합니다.
    return float(value)

# 반지름을 입력받아 원의 둘레를 계산하는 함수입니다.
def get_circum(radius):
    # 원의 둘레 계산 공식: 2 * π * 반지름
    return 2 * p1 * radius

# 반지름을 입력받아 원의 넓이를 계산하는 함수입니다.
def get_circle(radius):
    # 원의 넓이 계산 공식: π * 반지름^2
    return p1 * radius * radius
````
