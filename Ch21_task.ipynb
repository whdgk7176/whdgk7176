```python
def sum_numbers(*numbers):
    total = 0
    for n in numbers:
        total += n
    return total
print(sum_numbers(1,2,3,4,67))
```

[문제 6] 0부터 시작하여 n까지의 숫자 중에서 홀수만 생성하는 제너레이터 odd_numbers를 작성하세요.

```python
def odd_numbers(n):
    for i in range(n+1):
        if  i % 2 != 0:
            yield i

odd_numbers = (i for i in range(30) if  i % 2 !=0 )
for num in odd_numbers:
    print(num)
```

[문제 10] 팩토리얼을 계산하는 재귀 함수 factorial을 작성하세요.

```python
def factorial(n):
    if n == 0 :
        return 1
    return n * factorial(n-1)
factorial(5)
```

[문제 11] 문자열 리스트를 받아, 각 문자열의 길이를 출력하는 함수 print_lengths를 작성하세요.

```python
def print_lengths(strings):
    for w in strings:
        print(len(w))



words = ["apple", "banana", "watermelon", "muskmelon"]

print_lengths(words)
```

[문제 12] 두 문자열을 받아서 긴 문자열을 반환하는 함수 longer_string을 작성하세요. 만약 두 문자열의 길이가 같다면, 첫 번째 문자열을 반환하세요.

```python
def longer_string(x,y):
    if len(x) == len(y):
        return x
    elif len(x) > len (y):
        return x
    else:
        return y

print(longer_string("log", "apple"))
print(longer_string("winterdream", "cock"))
print(longer_string("oculus", "auriss"))
```
